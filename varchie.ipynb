```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
number_of_tweets = 15000
tweets = []
quote_status = []
lang = []
favorite_count = []
retweet_count = []
time = []
tweet_id = []

for i in tweepy.Cursor(api.search, q = "time jump Riverdale", tweet_mode="extended", wait_on_rate_limit="True").items(number_of_tweets):
    tweets.append(i.full_text)
    quote_status.append(i.is_quote_status)
    lang.append(i.lang)
    favorite_count.append(i.favorite_count)
    retweet_count.append(i.retweet_count)
    time.append(i.created_at)
    tweet_id.append(i.id)
```

```python
df = pd.DataFrame({'tweets': tweets, 'quote_status': quote_status, 'language': lang, 'favorite_count': favorite_count, 'retweet_count': retweet_count, 'time':time, "id":tweet_id})
```

```python
df
```

```python
df.drop_duplicates(subset='tweets', inplace=True)
```

```python

df
```

```python
df=df[df.language =="en"]
```

```python
df=df.reset_index(drop=True)
```
