```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
# set number of tweets
number_of_tweets = 15000

# create lists to store twitter data
tweets = []
quote_status = []
lang = []
favorite_count = []
retweet_count = []
time = []
tweet_id = []

# pull tweets and save relevant data to designated lists, change q for different search terms
for i in tweepy.Cursor(api.search, q = "riverdale", tweet_mode="extended", wait_on_rate_limit="True").items(number_of_tweets):
    tweets.append(i.full_text)
    quote_status.append(i.is_quote_status)
    lang.append(i.lang)
    favorite_count.append(i.favorite_count)
    retweet_count.append(i.retweet_count)
    time.append(i.created_at)
    tweet_id.append(i.id)
```

```python
# write lists to dataframe
df = pd.DataFrame({'tweets': tweets, 'quote_status': quote_status, 'language': lang, 'favorite_count': favorite_count, 'retweet_count': retweet_count, 'time':time, "id":tweet_id})
```

```python
# drop duplicate tweets
df.drop_duplicates(subset='tweets', inplace=True)
```

```python
#filter tweets to show english only
df=df[df.language =="en"]
```

```python
#reset index
df=df.reset_index(drop=True)
```
