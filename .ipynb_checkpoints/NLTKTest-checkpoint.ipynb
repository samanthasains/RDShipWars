```python
# Code for NLTK example
import string 
from collections import Counter

import pandas as pd
import matplotlib.pyplot as plt
import re
import emoji
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
choni_csv = pd.read_csv('choni.csv');
choni_csv.head(10)
```

```python
def sentiment_analysis(sentiment_text):
    score = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)
    print('---score:---')
    print(score)
    
def normalize_Text(tweet):
    print('---[START]text before modifying:---')
    print(tweet)
    tweet = tweet.lower()
    #text = emoji.demojize(tweet)
    #print('---emoji modified text---')
    #print(text)
    text = tweet
    text = re.sub("@[A-Za-z0-9]+","", text) #Remove @ sign
    text = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", text) #Remove http links
    text = text.replace("#", "").replace("_", " ") #Remove hashtag sign but keep the text
    print("---modified text---")
    print(text)
    sentiment_analysis(text)
    print("---analysis with emotes converted---")
    text = emoji.demojize(tweet)
    #print('---emoji modified text---')
    #print(text)
    sentiment_analysis(text)
    print('---[END]---')
```

```python
for ind in choni_csv.index:
    normalize_Text(choni_csv['tweets'][ind])
    #print(choni_csv['tweets'][ind])
```
